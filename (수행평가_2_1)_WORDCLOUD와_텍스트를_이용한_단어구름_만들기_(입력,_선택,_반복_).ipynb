```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

text = "머신러닝 K-Mean 군집 해석 해석 해석 해석 이상치 이상치 이상치 초기값 초기값 초기값 간단 간단 간단 쉬움 쉬움 쉬움" ###시각화 대상이 되는 단어들을 띄어쓰기로 구별하여 입력"
print(text) #위에서 입력한 것을 출력해줌 
word = WordCloud(font_path='/content/gdrive/MyDrive/1TEST(테스트용)/BMJUA_ttf.ttf', background_color='white').generate(text) ###worldcloud라는 변수(공간)에 WordCloud라는 도구를 사용하여 시각화 설정

plt.figure(figsize=(20,20)) #이미지 사이즈 지정 
plt.imshow(word, interpolation='lanczos') #wordcloud안에 들어있는 설정값을 이용하여 시각화를 하고 그 결과값을 저장, 뒤에 있는 것은 부드러움의 정도 (신경X)
plt.axis('off') #x y 축 숫자 제거
plt.savefig('/content/gdrive/MyDrive/1TEST(테스트용)/비지도.png') ###위에서 만든 결과값을 파일로 만듦 
plt.show()  #최종 결과를 그림으로 보여줌
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS #시각화를 위한 도구인 wordcloud(전체)에서 시각화를 담당하는 WordCloud도구와, 특정 단어를 제거할 수 있는 STOPWORDS 도구를 불러옴

f = open('/content/gdrive/MyDrive/1TEST(테스트용)/오타니.txt','r') ### 파일의 경로를 복사 후 붙여넣기
data = f.read()

spwords=set(STOPWORDS) # 기본적으로 많이 쓰는 단어 (제외할 단어) # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..
spwords.add('들어') ###제외하고 싶은 단어 추가 * 시각화에서 큰 의미가 없는 단어들을 최소 3개 이상 채우기
spwords.add('3회') ###제외하고 싶은 단어 추가
spwords.add('이날') ###제외하고 싶은 단어 추가
spwords.add('김주희') ###제외하고 싶은 단어 추가

word = WordCloud(max_font_size=200, font_path='/content/gdrive/MyDrive/1TEST(테스트용)/BMJUA_ttf.ttf',    ###폰트 크기, 폰트 경로, 제외할 단어, 배경색, 크기.generate(대상이 되는 글자)
                     stopwords=spwords,
                     colormap='cool', #위에서 골라서 바꿔도 됨 (필수는 아님)
                     background_color='#e6e6fa', #위에서 골라서 바꿔도 됨 (필수는 아님)
                     width=1200,height=800).generate(data)

plt.figure(figsize=(10,8)) #크기 지정 
plt.imshow(word) #이미지 생성 
plt.tight_layout(pad=0) #여백 설정(여백을 주고 싶으면 바꾸기)
plt.axis('off') #x,y축 값 삭제 
plt.savefig('/content/gdrive/MyDrive/1TEST(테스트용)/오타니.png') #파일로 생성(선택)
plt.show() #생성된 시각화 보여주기 
f.close() #파일 사용 마지막
```

### 여러 개의 텍스트를 반복하여 워드 클라우드로 만들고 저장하기

- **폰트 경로를 위에서 수행했던 걸 참고해서 복사 붙여넣기 하고** 
- **plt.savefig(' ') 제일 첫번째에 있는 작은 따옴표 사이에 정보-2021 경로를 복사 붙여넣기 하세요.**

```python
import requests #인터넷에 있는 파일 내용을 불러올 때 사용하는 도구

filelst = ["https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/ok.txt", #옥단춘전 정보 
           "https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/%EB%B2%84%ED%84%B0(BTS).txt", #BTS butter 가사
           "https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/%EB%A1%A4%EB%A6%B0(Rollin').txt", #브레이브걸스 Rollin' 가사 
           "https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/%EB%A0%88%EB%B9%84%EB%82%98%EC%8A%A4.txt", # 6월 모의고사 비문학 지문 (레비나스)
           "https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/%EA%B1%B0%EB%B6%81%EC%9D%B4(%EC%82%AC%EA%B3%84).txt"] # 거북이 사계 가사 

for i in range(len()): ###len은 크기를 의미함, 어떤 리스트의 크기만큼을 구해야 할까? i는 0부터 시작 
  txt = requests.get(filelst[]).text ###filelist의 O번째 값을 가져와 텍스트 파일로 만들어 저장해야 할까?
  #print(txt)

  word = WordCloud(max_font_size=200, font_path='',    ###폰트 경로 복사해서 붙여넣어주기
                     colormap='Spectral', #위에서 골라서 바꿔도 됨 (필수는 아님)
                     background_color='#fdf5e6', #위에서 골라서 바꿔도 됨 (필수는 아님)
                     width=1200,height=800).generate(txt)

  plt.figure(figsize=(10,8)) #크기 지정 
  plt.imshow(word) #이미지 생성 
  plt.tight_layout(pad=0) #여백 설정(여백을 주고 싶으면 바꾸기)
  plt.axis('off') #x,y축 값 삭제 
  plt.savefig(''+'/'+ str(i) +"번째 파일" +".png") ### 
  plt.show() #생성된 시각화 보여주기 
                
```
